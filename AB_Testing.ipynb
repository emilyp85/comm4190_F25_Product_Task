## Sample size & statistical test

I computed a sample-size estimate for a two-proportion test to detect an **absolute lift** of 2 percentage points (for example, baseline conversion 5% → 7%) with significance level alpha = 0.05 and power = 0.80.

Step-by-step arithmetic (summary):

* Let p1 = 0.05 (baseline conversion), p2 = 0.07 (target conversion), p = (p1 + p2)/2 = 0.06.
* Use z_alpha = 1.96 (two-sided 95% CI), z_beta = 0.84 (power 80%).
* Compute required sample per group using the standard approximation for two proportions:

  n ≈ [ z_alpha * sqrt(2 * p * (1 - p)) + z_beta * sqrt(p1*(1-p1) + p2*(1-p2)) ]^2 / (p2 - p1)^2

Plugging in numbers yields **n ≈ 2210** users per variant (that is, ~2210 visitors to Variant A and ~2210 to Variant B) to have a good chance of detecting a 2 percentage point difference.

Interpretation and practical notes:

* If your baseline conversion is higher (e.g., 10%), sample size required will be different (typically lower to detect the same absolute uplift).
* If you want to detect a smaller difference (e.g., 1% absolute), the sample size will grow substantially.
* The calculation above assumes independent visitors and a two-sided test. Use a one-sided test if you only care about uplift in one direction (reduces required sample a little).

## Statistical analysis method

* For the primary metric, use a two-proportion z-test to compare conversion rates (or Fisher’s exact test if counts are small).
* Report conversion rates, absolute difference, relative lift, 95% confidence interval, and p-value.
* Use Bayesian credible intervals if you prefer Bayesian A/B interpretation.
* Adjust for multiple comparisons if you run many simultaneous secondary tests (Bonferroni or, better, pre-register primary metric and treat other tests as exploratory).

## Qualitative testing & triangulation

Numbers tell whether something works; qualitative feedback tells **why**. Do these in parallel:

* 5–10 moderated user interviews or usability tests where participants think aloud while visiting each variant.
* Session replays & heatmaps to spot where users drop off or hesitate.
* Short exit survey on both pages asking “Why didn’t you buy?” or “What’s your impression?” (multiple choice + short text).

## Success criteria / decision rules

* If Variant B’s conversion rate is significantly higher (p < 0.05) and the effect size is meaningful (your chosen business threshold — e.g., >10% relative lift OR >1% absolute lift), deploy Variant B.
* If results are not significant but show a directional lift, iterate: test a high-impact element (hero, CTA, price).
* If Variant A outperforms B, analyze heatmaps and survey responses to see which messaging resonated and adopt those elements.

## Implementation checklist
